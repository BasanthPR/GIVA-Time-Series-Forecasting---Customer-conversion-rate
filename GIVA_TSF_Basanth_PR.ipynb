# Importing modules and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from pandas import Series
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
df=pd.DataFrame()
df=pd.read_excel('/content/Excel_Sample_BA.xlsx',sheet_name=0,index_col='Date',parse_dates=True)
df=df.dropna()
print(df)
```

```python
df.tail()
```

```python
df['Overall Conv'].plot(figsize=(12,5))
```

```python
df['Conv_LastMonth']=df['Overall Conv'].shift(+1)
df['Conv_2Monthsback']=df['Overall Conv'].shift(+2)
df['Conv_3Monthsback']=df['Overall Conv'].shift(+3)
df
```

```python
df=df.dropna()
df
```

# Importing Random Forest Regressor Algorithm

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,max_features=3, random_state=1)
```

```python
import numpy as np
x1,x2,x3,y=df['Conv_LastMonth'],df['Conv_2Monthsback'],df['Conv_3Monthsback'],df['Overall Conv']
x1,x2,x3,y=np.array(x1),np.array(x2),np.array(x3),np.array(y)
x1,x2,x3,y=x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),y.reshape(-1,1)
final_x=np.concatenate((x1,x2,x3),axis=1)
print(final_x)
```

# Splitting the original dataset into Training and Testing dataset

```python
X_train,X_test,y_train,y_test=final_x[:-30],final_x[-30:],y[:-30],y[-30:]
model.fit(X_train,y_train)
```

# Fitting testing data over RFR

```python
pred=model.predict(X_test)
plt.rcParams["figure.figsize"] = (12,8)
plt.plot(pred,label='Random_Forest_Predictions')
plt.plot(y_test,label='Actual Conv')
plt.legend(loc="upper left")
plt.show()
```

# Finding RMSE

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse_rf=sqrt(mean_squared_error(pred,y_test))
```

```python
print('Mean Squared Error for Random Forest Model is:',rmse_rf)
```
